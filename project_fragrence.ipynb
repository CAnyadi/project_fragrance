```python
#minimum votes
min= 'SELECT brand, MIN(votes) FROM df_perfume'
print(pysqldf(min))
```

```python
#most popular frangrance ordered by rating score 
title= 'SELECT brand,title,rating_score,votes FROM df_perfume GROUP BY brand ORDER BY rating_score DESC LIMIT 15'
print(pysqldf(title))
```

```python
# most popular fragrence by longevity so longevity_long or longevity_extreme > 100 order by longevity_very_long
longevity = 'SELECT brand, title ,rating_score, votes, longevity_poor,longevity_weak,longevity_long, longevity_very_long FROM df_perfume WHERE longevity_long > 100 OR longevity_very_long > 100 ORDER BY rating_score DESC LIMIT 15  '
print(pysqldf(longevity))
```

```python
#what are the rating for perfumes with the lowest longevity and longevity_long and very_long and < 100
low_longevity = 'SELECT brand, title, rating_score,votes,longevity_poor,longevity_weak, longevity_long, longevity_very_long FROM df_perfume WHERE (longevity_weak >100 OR longevity_poor > 100) AND(longevity_long <100 AND longevity_very_long <100) ORDER BY longevity_poor DESC LIMIT 20'
print(pysqldf(low_longevity))
```
